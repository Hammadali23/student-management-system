```python
class Student:
    def __init__(self, name, age, class_name):
        self.name = name
        self.age = age
        self.class_name = class_name

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Class: {self.class_name}"

def add_student(students):
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    class_name = input("Enter student class: ")
    student = Student(name, age, class_name)
    students.append(student)
    print(f"Student {name} added successfully!")

def display_students(students):
    if not students:
        print("No students found.")
    else:
        for student in students:
            print(student)

def save_to_file(students):
    with open("students.txt", "w") as file:
        for student in students:
            file.write(f"{student.name},{student.age},{student.class_name}\n")
    print("Data saved to file successfully.")

def load_from_file():
    students = []
    try:
        with open("students.txt", "r") as file:
            for line in file:
                name, age, class_name = line.strip().split(',')
                students.append(Student(name, age, class_name))
    except FileNotFoundError:
        print("File not found. Starting with an empty list.")
    return students

def search_student(students):
    name = input("Enter name to search: ")
    for student in students:
        if student.name.lower() == name.lower():
            print(student)
            return
    print("Student not found.")

def main():
    students = load_from_file()
    while True:
        print("\n1. Add Student")
        print("2. Display Students")
        print("3. Search Student")
        print("4. Save and Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            add_student(students)
        elif choice == "2":
            display_students(students)
        elif choice == "3":
            search_student(students)
        elif choice == "4":
            save_to_file(students)
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()


```
